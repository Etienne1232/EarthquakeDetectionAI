```python
combined_data = load('../DataDumps/SpectralSubt.joblib')
combined_labels = load('../DataDumps/labels.joblib')
import numpy as np
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
from keras.layers import LSTM, Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(combined_data, combined_labels, test_size=0.2, stratify=combined_labels, random_state=0)

X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, stratify=y_train, random_state=0)

def create_model(lstm_units=100, dropout_rate=0.2, learning_rate=0.001):
    model = Sequential()
    model.add(LSTM(units=lstm_units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(units=lstm_units))
    model.add(Dropout(dropout_rate))
    model.add(Dense(units=3, activation='softmax'))  # Adjust units here based on the number of classes

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create and compile the model
model = create_model()
checkpoint_callback = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True, mode='min', verbose=1)

# Train the model
history = model.fit(X_train, y_train, epochs=400, batch_size=32, validation_data=(X_dev, y_dev), callbacks=[checkpoint_callback])

best_epoch = np.argmin(history.history['val_loss'])

# Load the best model
model.load_weights('best_model.h5')

# Evaluate the best model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Best model: Loss={loss}, Accuracy={accuracy}')
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)

test_acc = accuracy_score(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
print('Test Accuracy:', test_acc)
```

```python
from joblib import load

combined_data = load('../DataDumps/SpectralSubt.joblib')
combined_labels = load('../DataDumps/labels.joblib')
combined_labels[combined_labels == 2] = 1


import numpy as np
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
from keras.layers import LSTM, Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(combined_data, combined_labels, test_size=0.2, stratify=combined_labels, random_state=0)

# Split the training set further into training and dev sets (80% train, 20% dev)
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, stratify=y_train, random_state=0)

def create_model(lstm_units=100, dropout_rate=0.2, learning_rate=0.001):
    model1 = Sequential()
    model1.add(LSTM(units=lstm_units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
    model1.add(Dropout(dropout_rate))
    model1.add(LSTM(units=lstm_units))
    model1.add(Dropout(dropout_rate))
    model1.add(Dense(units=1, activation='sigmoid'))  # Adjust units here based on the number of classes

    model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model1

# Create and compile the model
model1 = create_model()
checkpoint_callback = ModelCheckpoint('2best_model.h5', monitor='val_loss', save_best_only=True, mode='min', verbose=1)

# Train the model
history = model1.fit(X_train, y_train, epochs=400, batch_size=32, validation_data=(X_dev, y_dev), callbacks=[checkpoint_callback])

best_epoch = np.argmin(history.history['val_loss'])

# Load the best model
model1.load_weights('2best_model.h5')

# Evaluate the best model
loss, accuracy = model1.evaluate(X_test, y_test)
print(f'Best model: Loss={loss}, Accuracy={accuracy}')

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model1.predict(X_test)
y_pred = np.round(y_pred)

test_acc = accuracy_score(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Test Accuracy:', test_acc)
```
