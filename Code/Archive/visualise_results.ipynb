```python
from joblib import load
import obspy
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_set = load('../Dump2/Test_Data_Raw.joblib')
```

```python
def plot_data(test_set, classified_correctly):
    for idx, example in enumerate(test_set):
        if classified_correctly[idx]:
            continue
        fig, axs = plt.subplots(example.shape[0], sharex=True)
        axs[0].set_title("Wrong Classification")
        axs[0].plot(example[0])
        axs[0].set_ylabel("HHE")
        axs[1].plot(example[1])
        axs[1].set_ylabel("HHN")
        axs[2].plot(example[2])
        axs[2].set_ylabel("HHZ")

        plt.xlabel("Time") 
        plt.show()
```

```python
plot_data(test_set, [False, True])
```
